```python
def Age_category(X):
  if X <= 25:
    return 'Young'
  elif X >=26 and X <=35:
    return 'YoungAdult'
  elif X >=36 and X <= 45:
    return 'Adult'
  elif X >=46 and X <= 60:
    return 'SeniorAdult'
  else:
    return 'Elders'
```

```python
df['Age_Category']=df['Age'].apply(Age_category)
```

```python
def criteriat (X):
  if X >= 0.80:
    return 'Premium'
  elif X >=0.59 and X <=0.79:
    return 'Standard'
  elif X >=0.49 and X <=0.58:
    return 'Regular'
  elif X >=0.39 and X <=0.48:
    return 'Risk'
  elif X >=0.20 and X <=0.38:
    return 'Very_risky'
  elif X >=0.10 and X <=0.19:
    return 'Danger'
  else:
    return 'No Loan'
```

```python
df['Custumer Credit Scores']=df['Credit Utilization Ratio'].apply(criteriat)
```
